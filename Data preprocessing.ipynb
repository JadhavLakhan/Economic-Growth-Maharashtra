```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller

# Sample Data
data = {
    "Year": pd.date_range(start="2012-03-31", periods=12, freq="Y"),
    "GDDP": [1.280369e+06, 1.459629e+06, 1.649647e+06, 1.779138e+06, 
             1.966225e+06, 2.198185e+06, 2.352782e+06, 2.528854e+06, 
             2.656806e+06, 2.610651e+06, 3.144138e+06, 3.645884e+06]
}

df = pd.DataFrame(data).set_index("Year")

# Plot ACF & PACF
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sm.graphics.tsa.plot_acf(df["GDDP"], lags=6, ax=axes[0])
axes[0].set_title("Autocorrelation Function (ACF)")

sm.graphics.tsa.plot_pacf(df["GDDP"], lags=3, ax=axes[1])
axes[1].set_title("Partial Autocorrelation Function (PACF)")

plt.show()
```
